```python
# Search for monk parakeet from 1970-2015 in Florida

# import dependencies
import os
import requests
import json
import pandas as pd
import plotly.express as px
from config import api_key
from config import mapbox_token
from calendar import monthrange

%reload_ext lab_black

px.set_mapbox_access_token(mapbox_token)
```

```python
# Sandbox for trial & error
df_empty = pd.DataFrame([])
bool("speciesCode" in df_empty.columns.tolist())
```

```python
column_headers = [
    "speciesCode",
    "comName",
    "sciName",
    "locId",
    "locName",
    "obsDt",
    "howMany",
    "lat",
    "lng",
    "obsValid",
    "obsReviewed",
    "locationPrivate",
    "subId",
]

parakeets_df = pd.DataFrame(columns=column_headers)

for year in range(1970, 2016):
    for month in range(1, 13):
        for day in range(1, monthrange(year, month)[1] + 1):
            url = (
                f"https://api.ebird.org/v2/data/obs/US-FL/historic/{year}/{month}/{day}"
            )
            r = requests.get(url, params={"key": api_key})
            data = r.json()
            df = pd.DataFrame(data)
            if "speciesCode" in df.columns.tolist() and bool(
                "monpar" in df["speciesCode"].to_list()
            ):
                parakeets_df = pd.merge(
                    parakeets_df, df.loc[(df["speciesCode"] == "monpar")], how="outer"
                )
                print(
                    f"Monk Parakeet seen on {month}/{day}/{year} in Florida!!!!!!!!!!!!!!!!!"
                )
            else:
                print(f"No monk parakeets in Florida on {month}/{day}/{year}")
        try:
            parakeets_df.to_csv("florida_parakeets.csv", index=False)
        except Exception as e:
            print(f"No monk parakeet data in Florida to save yet.")
            print(e)

parakeets_df
```
