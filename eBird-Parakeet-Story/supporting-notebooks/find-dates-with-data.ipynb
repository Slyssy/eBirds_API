```python
# Searches eBird historic API for dates with bird observations for years 1800-1999

# import dependencies
import os
import requests
import json
import pandas as pd
import plotly.express as px
from config import api_key
from config import mapbox_token
from calendar import monthrange

%reload_ext lab_black

px.set_mapbox_access_token(mapbox_token)
```

```python
product_column_names = [
    "profileHandle",
    "userDisplayName",
    "numSpecies",
    "numCompleteChecklists",
    "rowNum",
    "userId",
]
historic_birds_df = pd.DataFrame(columns=product_column_names)

dates_with_data = {}

for year in range(1800, 2000):
    for month in range(1, 13):
        for day in range(1, monthrange(year, month)[1] + 1):
            url = f"https://api.ebird.org/v2/product/top100/US/{year}/{month}/{day}"
            r = requests.get(url, params={"key": api_key})
            current_df = pd.DataFrame(r.json())
            if not current_df.empty:
                pd.concat([historic_birds_df, current_df])
                dates_with_data.setdefault("year", []).append(year)
                dates_with_data.setdefault("month", []).append(month)
                dates_with_data.setdefault("day", []).append(day)
                dates_with_data.setdefault("num_products", []).append(len(r.json()))
                print(f"Date {month}/{day}/{year} has data!!!!!!!!!!")
            else:
                print(f"No data for {month}/{day}/{year}")

df = pd.DataFrame(dates_with_data)
df.to_csv("dates-with-data.csv", index=False)
```
