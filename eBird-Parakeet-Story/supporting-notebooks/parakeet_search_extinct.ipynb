```python
# Asks eBird API for "product status" of multiple years -- returns number of checklists, contributors, and species logged for requested dates
# Iterates over years 1800-1999

# import dependencies
import os
import requests
import json
import pandas as pd
import plotly.express as px
from config import api_key
from config import mapbox_token
from calendar import monthrange

%reload_ext lab_black

px.set_mapbox_access_token(mapbox_token)
```

```python
carolina_parakeet_range = ["TN", "VA", "NC", "SC", "GA", "FL", "AL", "MS", "LA", "IA", "AR", "KY", "WV"]

historic_dates_df = pd.read_csv("bird_dates_all.csv")
search_dates_df = historic_dates_df.loc[(historic_dates_df["year"] <= 1939) & (historic_dates_df["species"] > 0)][
    ["year", "month", "day"]
]

column_headers = [
    "speciesCode",
    "comName",
    "sciName",
    "locId",
    "locName",
    "obsDt",
    "howMany",
    "lat",
    "lng",
    "obsValid",
    "obsReviewed",
    "locationPrivate",
    "subId",
]

parakeets_df = pd.DataFrame(columns=column_headers)

for date in search_dates_df.iterrows():
    year = date[1][0]
    month = date[1][1]
    day = date[1][2]
    for state in carolina_parakeet_range:
        url = f"https://api.ebird.org/v2/data/obs/US-{state}/historic/{year}/{month}/{day}"
        r = requests.get(url, params={"key": api_key})
        data = r.json()
        df = pd.DataFrame(data)
        if "speciesCode" in df.columns.tolist() and bool("carpar" in df["speciesCode"].to_list()):
            parakeets_df = pd.merge(parakeets_df, df.loc[(df["speciesCode"] == "carpar")], how="outer")
            print(f"Carolina Parakeet seen on {month}/{day}/{year} in {state}!!!!!!!!!!!!!!!!!")
        else:
                print(f"No parakeets on {month}/{day}/{year} in {state}")
    try:
        parakeets_df.to_csv("carolina_parakeets.csv", index=False)
    except Exception as e:
        print(f"No carolina parakeet data to save yet.")
        print(e)

parakeets_df
```
